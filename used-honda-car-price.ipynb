## Changing Column Data Type

### Creating new column

```python
#Price data not in numeric, so creating new column
work_data['Price (Lakh)']=work_data['Price'].str.replace(' Lakh','').str.strip()
```

```python
work_data.head()
```

### Changing data type into float

```python
work_data['Price (Lakh)']=work_data['Price (Lakh)'].str.replace(',','').astype(float)
```

```python
work_data.dtypes
```

### Removing Price column

```python
work_data=work_data.drop('Price',axis=1)
```
